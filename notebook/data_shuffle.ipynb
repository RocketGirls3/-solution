```python
import pandas as pd
import numpy as np
import random
from tqdm import tqdm
```

```python
train = pd.read_csv('../input/new_data/train_set.csv')
test = pd.read_csv('../input/new_data/test_set.csv')
```

```python
label_1_text = train.loc[train['class']==1]['word_seg'].values
```

```python
shuffle_1_data = []
shuffle_19_data = []
```

```python
for i in tqdm(label_1_text):
    words = i.split(' ')
    rs=random.sample(words,len(words))
    rs = ' '.join(rs)
    shuffle_1_data.append(rs)
```

```python
label_19_text = train.loc[train['class']==19]['word_seg'].values
```

```python
for i in tqdm(label_19_text):
    words = i.split(' ')
    rs=random.sample(words,len(words))
    rs = ' '.join(rs)
    shuffle_19_data.append(rs)
```

```python
df_label19 = pd.DataFrame(shuffle_19_data,columns=['word_seg'])
df_label19['class'] = 19
df_label19['article']='0 0 0 0'
```

```python
df_label1 = pd.DataFrame(shuffle_1_data,columns=['word_seg'])
df_label1['class'] = 1
df_label1['article']='0 0 0 0'
```

```python
shuffle_train = pd.concat([train, df_label19, df_label1])
shuffle_train = shuffle_train.reset_index(drop=True)
```

```python
shuffle_train.to_csv('../input/new_data/shuffle_train.csv',index=False)
```

```python
SAMPLE_RATE = 0.15
SHUFFLE_RATE = 0.3
def sample_enhance(x,sample_rate):
    """
    sample index for enhance, 1/5 data every time
    """
    shuffle_idx = []
    for i in range(0,len(x)):
        shuffle_idx.append(random.randint(0,len(x)-1))
    return list(set(shuffle_idx[:int(len(shuffle_idx)*sample_rate)]))


def shuffle_comment(comment,shuffle_rate):
    try:
        word_list = comment.split(' ')
        length = len(word_list)
        shuffle_len = int(length*shuffle_rate)
        start = random.randint(0,length-shuffle_len-1)
        a = word_list[start:start+shuffle_len]
        random.shuffle(a)
        word_list[start:start + shuffle_len] = a
        comment = " ".join(word_list)
    except:
        print(comment)
        return comment
    return comment
```

```python
idx = sample_enhance(train,SAMPLE_RATE)
```

```python
# idx = [0,1,2,3,4,5]
sample_train = train.iloc[idx]
```

```python
sample_train['word_seg'] = sample_train['word_seg'].map(lambda x:shuffle_comment(x,SHUFFLE_RATE))
```
